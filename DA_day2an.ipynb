```python
#show the core compresion of 2students in 5 diff sub as a line plot sub as x axis and marks as y axis
s1=np.array[88,66,44,33,22]
s2=[12,45,67,23,77]
sub=["t","h","e","m","s"]
plt.subplot(2,1,1)#1 graph divide it as 2:parts,1:upper part   #subplot as a keyword
plt.plot(sub,s1,label="s1",color="red")
plt.legend()#continuous line
plt.subplot(2,1,2)#1 graph divide it as 2:parts,1:lower part   #subplot as a keyword
plt.plot(sub,s2,label="s2",color="green")
plt.legend()#continuous line
```
